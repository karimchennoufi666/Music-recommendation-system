```python
import pandas as pd
import numpy as np

def get_mean_vector(song_list):
    """
    Calculate the mean vector of numerical features for a list of songs.

    Parameters:
    - song_list (list of str): List of song names.

    Returns:
    - pd.Series or None: Mean vector of numerical features, or None if no songs are found.
    """
    # Filter dataset for the songs in the list
    filtered_songs = data[data['name'].str.lower().isin([s.lower() for s in song_list])]
    
    if filtered_songs.empty:
        return None  # No songs found
    
    # Select only numerical columns
    numerical_cols = filtered_songs.select_dtypes(include=np.number)
    
    # Calculate mean vector
    mean_vector = numerical_cols.mean()
    
    return mean_vector
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

client_id = "f7f30d8beeac46aa9e63c4fb644005ed"
client_secret = "d64f8e38e527458191da4b9e15c155bb"


sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=client_id, client_secret=client_secret))

# Your datasets
# gons_clusters = pd.read_csv("/path/to/gons_clusters.csv")  # your main dataset
# spotify_track = pd.read_csv("/path/to/spotify_track.csv")  # your track list

# Function to get song data from dataset or Spotify
def get_song_data(song, dataset):
    row = dataset[dataset["name"].str.lower() == song.lower()]
    if not row.empty:
        return row.iloc[0].to_dict()
    try:
        results = sp.search(q=f"track:{song}", type="track", limit=1)
        tracks = results["tracks"]["items"]
        if tracks:
            track = tracks[0]
            return {
                "name": track["name"],
                "artists": [a["name"] for a in track["artists"]],
                "id": track["id"],
                "popularity": track["popularity"],
                "album": track["album"]["name"],
                "release_date": track["album"]["release_date"],
            }
    except:
        return None
    return None

# Compute mean vector of features for a list of songs
def get_mean_vector(song_list, dataset, feature_cols):
    vectors = []
    for song in song_list:
        data = get_song_data(song, dataset)
        if data:
            vec = []
            for col in feature_cols:
                value = data.get(col)
                if value is None or not isinstance(value, (int, float)):
                    vec.append(0)  # or np.nan, but 0 is safer for similarity
                else:
                    vec.append(float(value))
            vectors.append(vec)
    if not vectors:
        return None
    return np.mean(vectors, axis=0)

# Flatten a list of dictionaries
def flatten_dict_list(dict_list):
    flattened = defaultdict(list)
    for d in dict_list:
        for key, value in d.items():
            flattened[key].append(value)
    return dict(flattened)

# Recommend songs based on seed list
def recommend_songs(seed_songs, dataset, feature_cols, n=5):
    mean_vector = get_mean_vector(seed_songs, dataset, feature_cols)
    if mean_vector is None:
        return pd.DataFrame(columns=["name", "artists", "similarity"])

    # Use the dataset argument
    song_features = dataset[feature_cols].values
    similarities = cosine_similarity([mean_vector], song_features)[0]

    dataset["similarity"] = similarities
    recommendations = dataset.sort_values("similarity", ascending=False)
    recommendations = recommendations[~dataset["name"].str.lower().isin([s.lower() for s in seed_songs])]

    return recommendations.head(n)[["name", "artists", "similarity"]]

# Define numeric features present in your 'data' DataFrame
feature_columns = ["valence", "danceability", "energy", "loudness", "acousticness", 
                   "instrumentalness", "liveness", "speechiness", "tempo"]

# Call the recommendation system with your dataset
recs = recommend_songs(spotify_tracks["name"].tolist(), data, feature_columns, n=5)
print(recs)
```
